```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Sıra ile tüm img dimensionların en iyi modelleri çağırılıp teste giricek
```

```python
best_cnn_64x64 = "best_model_64x64.keras"
ımg_dimensions= (64, 64)
batch_sz = 32
```

```python
model_1 = load_model(best_cnn_64x64)

# Kendi kodlarımdaki gibi normalize ediyorum
test_gen = ImageDataGenerator(rescale=1. / 255)

#Sayın Hocam Test dosyası ve Data_Test klasörü aynı klasörde olmalı
test_data = test_gen.flow_from_directory(
    "Data_Test/",
    target_size=ımg_dimensions,
    batch_size=batch_sz,
    class_mode='categorical',
    shuffle=False  
)
```

```python
# Doğru labellar
#Bahsettiğim uyarı burada geliyor bu sadece bir uyarı kodun işleyişini ve çalışmasını etkilemiyor
actual_labels = test_data.classes
label_names = list(test_data.class_indices.keys()) 

#
predicted_probs = model_1.predict(test_data)


predicted_labels = np.argmax(predicted_probs, axis=1)

#Sayın Hocam burada report veriyoruz böylece sonuçları görebileceğiz
print("=== Classification Report ===")
print(classification_report(actual_labels, predicted_labels, target_names=label_names))

# Burası ve üst kısım dediğim gibi tamamen kendi kodumla aynı
conf_matrix = confusion_matrix(actual_labels, predicted_labels)

plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=label_names,
    yticklabels=label_names
)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - 64x64 Model")
plt.tight_layout()

plt.show()
```
